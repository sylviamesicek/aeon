```python
import os
from pathlib import Path

import ipynbname

primary_path = Path(ipynbname.path()).parent.parent.parent
print("Changing working directory to: ", primary_path)
os.chdir(primary_path)

import glob

import numpy as np
import pyvista as pv

pv.global_theme.font.family = 'times'

pv.global_theme.axes.x_color = 'black'
pv.global_theme.axes.y_color = 'black'
pv.global_theme.axes.z_color = 'black'
pv.global_theme.colorbar_orientation = 'vertical'
pv.global_theme.allow_empty_mesh = True
pv.global_theme.font.color = 'black'

ncontours = 100
# ncontours = 30
```

```python
plotter = pv.Plotter(notebook=False, off_screen=True)
# Open a gif
plotter.open_gif("figures/axiprolate_collapse.gif", fps=24.0)
# plotter.open_movie("figures/axiprolate_collapse.mp4", framerate=24)
screenshot = False

nfiles = len(glob.glob("output/axiprolate/0.4/evolve/*.vtu"))
for i in range(0, nfiles):
    if i % 2 != 0:
        continue

    file = f"output/axiprolate/0.4/evolve/axiprolate_{i}.vtu"

    plotter.clear()

    mesh = pv.read(file)    
    contour = mesh.contour(isosurfaces=np.linspace(-0.6, 0.4, ncontours), scalars="Field::Phi0", method="contour").translate([0, 0, 1.0])
    plotter.add_mesh(
        mesh,
        scalars="Field::Phi0",
        lighting=False,
        scalar_bar_args={"title": "Phi", "vertical": True, "position_x": 0.89, "position_y": 0.28, "color": "black" },
        clim=[-0.6, 0.4]
    )
    plotter.add_mesh(contour, lighting=False, color="black", line_width=2, point_size=0)
    plotter.background_color = 'white'
    plotter.add_title('Collapsing Prolate Scalar Field', color='black', font_size=14)
    axes = pv.CubeAxesActor(camera=plotter.camera, x_title="r", y_title="z", minor_ticks=True)
    axes.x_axis_range = (0, 48)
    axes.y_axis_range = (0, 48)
    axes.n_xlabels = 4
    axes.n_ylabels = 4
    axes.bounds = mesh.bounds 
    axes.use_2d_mode = True
    axes.GetXAxesLinesProperty().SetColor(0.0, 0.0, 0.0)
    axes.GetYAxesLinesProperty().SetColor(0.0, 0.0, 0.0)
    axes.GetXAxesLabelProperty().SetColor(0.0, 0.0, 0.0)
    axes.GetYAxesLabelProperty().SetColor(0.0, 0.0, 0.0)
    axes.GetXAxesTitleProperty().SetColor(0.0, 0.0, 0.0)
    axes.GetYAxesTitleProperty().SetColor(0.0, 0.0, 0.0)

    # print(isinstance(axes, vtk.vtkCubeAxesActor))
    actor, prop = plotter.add_actor(axes, render=False)
    # Write a frame. This triggers a render.
    plotter.camera.tight(view="xy", padding=0.4)
    plotter.write_frame()

    if float(i) / float(nfiles) > 8.0/8.2 and not screenshot:
        screenshot = True
        plotter.screenshot("figures/axiprolate_collapse.png")

plotter.close()
```

```python
plotter = pv.Plotter(notebook=False, off_screen=True)
# Open a gif
plotter.open_movie("figures/axiprolate_disperse_grid.mp4", framerate=24)

nfiles = len(glob.glob("output/axiprolate/0.3/evolve/*.vtu"))
for i in range(0, nfiles):
    file = f"output/axiprolate/0.3/evolve/axiprolate_{i}.vtu"

    if i > 400:
        continue

    plotter.clear()

    mesh = pv.read(file)
    plotter.add_mesh(
        mesh,
        scalars="Field::Phi0",
        lighting=False,
        scalar_bar_args={"title": "Phi", "vertical": True, "position_x": 0.89, "position_y": 0.28, "color": "white" },
        clim=[-0.3, 0.3],
        show_edges=True
    )
    plotter.background_color = 'black'
    plotter.add_title('Adaptive Mesh Refinement', color='white', font_size=14)
    axes = pv.CubeAxesActor(camera=plotter.camera, x_title="r", y_title="z", minor_ticks=True)
    axes.bounds = mesh.bounds 
    axes.use_2d_mode = True
    axes.GetXAxesLinesProperty().SetColor(1.0, 1.0, 1.0)
    # print(isinstance(axes, vtk.vtkCubeAxesActor))
    actor, prop = plotter.add_actor(axes, render=False)
    # Write a frame. This triggers a render.
    plotter.camera.tight(view="xy", padding=0.4)
    plotter.write_frame()

plotter.close()
```
