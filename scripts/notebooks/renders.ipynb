```python
import os
import glob
from pathlib import Path

import ipynbname

primary_path = Path(ipynbname.path()).parent.parent.parent
print("Changing working directory to: ", primary_path)
os.chdir(primary_path)

import numpy as np
import pyvista as pv

pv.global_theme.axes.x_color = 'black'
pv.global_theme.axes.y_color = 'black'
pv.global_theme.axes.z_color = 'black'
pv.global_theme.colorbar_orientation = 'vertical'
pv.global_theme.allow_empty_mesh = True
pv.global_theme.font.color = 'black'
```

```python
plotter = pv.Plotter(notebook=False, off_screen=True)
plotter.clear()

file = f"output/axicombo/0.0/0.3/initial/axicombo.vtu"
mesh = pv.read(file)
plotter.add_mesh(
    mesh, 
    scalars="IntField::BlockId", 
    lighting=False,
    show_scalar_bar=False,
    # scalar_bar_args={"title": "Phi", "vertical": True, "position_x": 0.89, "position_y": 0.28, "color": "black" },
    show_edges=True,
)
plotter.background_color = 'white'
# plotter.show_bounds(
#     bounds=mesh.bounds,
#     axes_ranges=(0.0, 48.0, 0.0, 48.0, 0.0, 1.0),
#     use_2d=True,
# )

# # plotter.show_bounds()
# # plotter.add_title('Collapsing Prolate Scalar Field', color='black', font_size=14)
print(mesh.bounds)
axes = pv.CubeAxesActor(camera=plotter.camera, x_title="r", y_title="z", minor_ticks=True)
# axes.bounds = mesh.bounds
axes.bounds = pv.BoundsTuple(x_min=0.0, x_max=48.0, y_min=0.0, y_max=48.0, z_min=0.0, z_max=0.0)
axes.x_axis_range = (0, 48)
axes.y_axis_range = (0, 48)
axes.n_xlabels = 4
axes.n_ylabels = 4
axes.use_2d_mode = True
axes.GetXAxesLinesProperty().SetColor(0.0, 0.0, 0.0)
axes.GetYAxesLinesProperty().SetColor(0.0, 0.0, 0.0)
axes.GetXAxesLabelProperty().SetColor(0.0, 0.0, 0.0)
axes.GetYAxesLabelProperty().SetColor(0.0, 0.0, 0.0)
axes.GetXAxesTitleProperty().SetColor(0.0, 0.0, 0.0)
axes.GetYAxesTitleProperty().SetColor(0.0, 0.0, 0.0)
# # print(isinstance(axes, vtk.vtkCubeAxesActor))
actor, prop = plotter.add_actor(axes, render=False)
# Write a frame. This triggers a render.
plotter.camera.tight(view="xy", padding=0.1)
plotter.render()
plotter.save_graphic("figures/block_id.pdf")
plotter.show()
```
